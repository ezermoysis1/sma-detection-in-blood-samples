```python
import matplotlib.pyplot as plt
from datetime import datetime
import matplotlib.dates as mdates

# Data
x = [datetime(2023, 6, 26), datetime(2023, 6, 28), datetime(2023, 6, 30),datetime(2023, 7, 2)]  # Dates
y = [50, 70, 80, 80]  # Accuracy
labels = ['Random classifier', 'Changed this', 'Changed that', 'Changed so']  # Text labels for each point
colors = ['red', 'green', 'blue', 'orange']  # Colors for each point

# Create the plot
for i in range(len(x)):
    plt.plot(x[i], y[i], marker='o', markersize=5, color=colors[i])  # Plot each point with different color

# Add a line passing through all points
plt.plot(x, y, 'k-')  # 'k-' specifies a black line passing through the points

plt.xlabel('Date')
plt.ylabel('Accuracy (%)')
plt.title('Project Progress')

# Set the x-axis limits
plt.xlim([datetime(2023, 6, 25), datetime(2023, 9, 10)])  # Adjust the x-axis limits as needed

# Set the y-axis limits
plt.ylim([0, 100])  # Adjust the y-axis limits as needed

# Format the x-axis ticks as dd-mmm-yy
date_format = mdates.DateFormatter('%d-%b-%y')  # Use '%d-%b-%y' for dd-mmm-yy format
plt.gca().xaxis.set_major_formatter(date_format)

# Rotate the x-axis labels
plt.tick_params(axis='x', rotation=90)  # Set rotation angle to 90 degrees

# Add text labels on the right side of each point with more space
for i in range(len(x)):
    plt.text(x[i], y[i], '    ' + labels[i], ha='left', va='center', fontweight='bold', fontsize=8, color=colors[i])

# Display the plot
plt.show()
```
