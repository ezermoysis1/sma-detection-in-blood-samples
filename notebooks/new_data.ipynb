```python
import os

def get_folder_names(directory_path):
    """Return a list of folder names in the given directory."""
    return [folder for folder in os.listdir(directory_path) if os.path.isdir(os.path.join(directory_path, folder))]

# Use the function
directory_path = 'edof_new_sma_2023'  # Replace with the desired path
folders = get_folder_names(directory_path)
print(folders)
```

```python
print(len(folders))
```

```python
import pandas as pd

# Define the file path and sheet name
file_path = 'slidedb_202306011328_FM10_FM20_v5-import_20230611-shared-print.xlsx'  # Assuming it's an Excel file
sheet_name = 'FM01-FM10-FM20_V4-2019'

# Read the specific sheet into a dataframe
df = pd.read_excel(file_path, sheet_name=sheet_name)

# Display the dataframe
print(df)
```

```python
# Define the list of columns to keep
columns_to_keep = ['FASt-Mal', 'Malaria', 'TYPE', 'SMA', 'PCV']

# Keep only the specified columns
df = df[columns_to_keep]

# Display the updated dataframe
print(df)
```

```python
df = df.rename(columns={'FASt-Mal': 'FASt-Mal-Code'})
df = df.rename(columns={'Malaria': 'MPDx_Pos-Neg-Null'})
df = df.rename(columns={'TYPE': 'ANAEMIA_SEVERITY'})
```

```python
df
```

```python
def classify(row):
    if row['MPDx_Pos-Neg-Null'] == 'Neg' and row['ANAEMIA_SEVERITY'] in ['NO', '[null]']:
        return 'No Malaria, No Anaemia'
    elif row['MPDx_Pos-Neg-Null'] == 'Pos' and row['ANAEMIA_SEVERITY'] in ['NO', '[null]']:
        return 'Malaria, No Anaemia'
    elif row['MPDx_Pos-Neg-Null'] == 'Pos' and row['ANAEMIA_SEVERITY'] in ['MILD', 'MODERATE']:
        return 'Malaria, Anaemia, No Severe'
    elif row['MPDx_Pos-Neg-Null'] == 'Pos' and row['ANAEMIA_SEVERITY'] == 'SEVERE':
        return 'Severe Malaria Anaemia'
    elif row['MPDx_Pos-Neg-Null'] == 'Neg' and row['ANAEMIA_SEVERITY'] in ['MILD', 'MODERATE']:
        return 'No Malaria, Anaemia'
    elif row['MPDx_Pos-Neg-Null'] == 'Neg' and row['ANAEMIA_SEVERITY'] == 'SEVERE':
        return 'No Malaria, Severe Anaemia'
    else:
        return 'Unclassified'

df['Diagnosis'] = df.apply(lambda row: classify(row), axis=1)
```
