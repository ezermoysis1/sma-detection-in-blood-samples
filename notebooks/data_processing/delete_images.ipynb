```python
import csv

loaded_list=[]
# Name of the CSV file
filename = 'images_to_remove1.csv'

# Creating an empty list to store the values
loaded_list = []

# Reading from the CSV
with open(filename, 'r') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        # Each row is a list, and we want the first (and only) element
        loaded_list.append(row[0])

# Printing the loaded list
print(loaded_list)
```

```python
loaded_list = [item.replace('edof_new_sma_2023_cells/', 'edof_new_sma_2023_cells_cleaned/') for item in loaded_list]
```

```python
print(loaded_list)
```

```python
# The list of file paths to be deleted
#files_to_delete = [s.replace('Red_Cell_Morpholog', 'clean_no_moderate_mild_20filter_balanced copy') for s in loaded_list]

files_to_delete=loaded_list
```

```python
import os
import shutil

for root, dirs, files in os.walk('edof_new_sma_2023_cells_cleaned'):
    for file in files:
        file_path = os.path.join(root, file)
        if file_path in files_to_delete:
            try:
                os.remove(file_path)
                print(f"File {file_path} has been deleted")
            except Exception as e:
                print(f"Error occurred while deleting file {file_path}: {e}")
```
